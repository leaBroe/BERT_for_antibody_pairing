```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statistics
import csv
```

# 100 pident CDRH3 paired

```python
# Load the data
paired_cdrh3_90 = pd.read_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/mmseqs/1.cdrh3/paired_rowid_cdrh3_100_clu_rep_idseq', header=None)

# Calculate CDR-H3 lengths
cdr3_aa_len_90 = paired_cdrh3_90[1].apply(lambda x: len(str(x)))

# Create a figure with 2 subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Boxplot on the first subplot
axs[0].boxplot(cdr3_aa_len_90)
axs[0].set_ylabel('CDR-H3 length')
axs[0].set_xlabel('CDR-H3')
axs[0].set_title('CDR-H3 - all human paired - 100 PIDENT')
text = f'# centroids seq: {len(paired_cdrh3_90)}'
axs[0].text(0.23, 0.9, text, transform=axs[0].transAxes, ha='center', va='center',
            bbox=dict(facecolor='white', edgecolor='black'))
text = f'AVG length: {round(statistics.mean(cdr3_aa_len_90))}'
axs[0].text(0.23, 0.80, text, transform=axs[0].transAxes, ha='center', va='center',
            bbox=dict(facecolor='white', edgecolor='black'))

# KDE plot on the second subplot
sns.histplot(cdr3_aa_len_90, kde=True, bins=100, ax=axs[1])
axs[1].set_xlabel('CDR-H3 length')
axs[1].set_ylabel('Frequency')
axs[1].set_title('CDR-H3 - all human paired - 100 PIDENT')

plt.tight_layout()

#plt.savefig('/ibmm_data2/oas_database/paired_lea_tmp/EDA_plots/clustering_cdrh3_all_default_90.png')

plt.show()
```
