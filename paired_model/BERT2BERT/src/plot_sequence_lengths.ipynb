```python
import matplotlib.pyplot as plt

# Initialize lists to store the lengths
heavy_lengths = []
light_lengths = []

# Open the file containing the sequences
with open('paired_full_seqs_sep_train_no_ids_space_separated.txt', 'r') as file:
    for line in file:
        if line.strip():  # Check if line is not empty
            parts = line.split('[SEP]')
            heavy_chain = parts[0].strip().replace(" ", "")  # Remove spaces
            light_chain = parts[1].strip().replace(" ", "")  # Remove spaces
            heavy_lengths.append(len(heavy_chain))
            light_lengths.append(len(light_chain))
```

```python
import numpy as np

plt.figure(figsize=(12, 8))

plt.scatter(light_lengths, heavy_lengths, s=30, alpha=0.4, edgecolors='w', linewidth=0.5)
plt.title('Distribution of Heavy and Light Chain Lengths', fontsize=16)
plt.xlabel('Length of Light Chain', fontsize=14)
plt.ylabel('Length of Heavy Chain', fontsize=14)
plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()

plt.show()
```
