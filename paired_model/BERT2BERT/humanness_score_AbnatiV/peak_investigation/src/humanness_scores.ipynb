```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statistics
import csv
```

```python
low_60_igk_gen_seqs = "/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/humanness_score_AbnatiV/peak_investigation/abnativ_results/igk_lower_60_generated_sequences_dna_anarci_aho_full_abnativ_seq_scores.csv"

low_60_igk_true_seqs = "/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/humanness_score_AbnatiV/peak_investigation/abnativ_results/igk_lower_60_true_sequences_dna_anarci_aho_full_abnativ_seq_scores.csv"

with open(low_60_igk_gen_seqs, "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
        break
```

```python
low_60_igk_gen_seqs_df = pd.read_csv(low_60_igk_gen_seqs)
low_60_igk_true_seqs_df = pd.read_csv(low_60_igk_true_seqs)

# extract AbNatiV VKappa Score
low_60_abnativ_vkappa_score_gen = low_60_igk_gen_seqs_df["AbNatiV VKappa Score"]
low_60_abnativ_vkappa_score_true = low_60_igk_true_seqs_df["AbNatiV VKappa Score"]


# plot histogram
sns.kdeplot(low_60_abnativ_vkappa_score_gen, alpha=0.5, color="orange")
sns.kdeplot(low_60_abnativ_vkappa_score_true, alpha=0.5, color="blue")
plt.legend(["Generated", "True"])
plt.xlabel("AbNatiV VKappa Humanness Score")
plt.ylabel("Frequency")
plt.title("Distribution of AbNatiV VKappa Humanness Score < 60% similarity")

print("Mean AbNatiV VKappa Score (Generated):", statistics.mean(low_60_abnativ_vkappa_score_gen))
print("Mean AbNatiV VKappa Score (True):", statistics.mean(low_60_abnativ_vkappa_score_true))

print("Median AbNatiV VKappa Score (Generated):", statistics.median(low_60_abnativ_vkappa_score_gen))
print("Median AbNatiV VKappa Score (True):", statistics.median(low_60_abnativ_vkappa_score_true))
```
