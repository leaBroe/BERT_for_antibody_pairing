```python
# env: OAS_paired_env
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# used model: 
# run name: full_diverse_beam_search_2_temp_0.5_max_length_150_early_stopping_true_batch_size_64_epochs_10_lr_0.0001_wd_0.1
# run name = name on wandb: https://wandb.ai/ibmm-unibe-ch/heavy2light_translation?nw=nwuserlea_broe
# reference name: h2l_div_beam_search_2_epoch_10_lr_1e-4_wd_0.1 (for directory names)
# metrics:
# # on full test set (67209 seqs):
# Average BLOSUM Score: 108.25002603817941
# Average Similarity Percentage: 33.05800137172086%
# Mean Perplexity: 2.1675915718078613
```

```python

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/sqlite3_data_for_analysis/evaluate_test_set_by_regions/h2l_div_beam_search_2_epoch_10_lr_1e-4_wd_0.1/fwr1_similarity_blosum_scores.csv')
```

```python
# Create a figure and axis objects for the plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the distribution of similarity scores
axes[0].hist(df['similarity_scores_fwr1'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_title('Distribution of Similarity Scores for FWR1')
axes[0].set_xlabel('Similarity Scores')
axes[0].set_ylabel('Frequency')

# Plot the distribution of blosum scores
axes[1].hist(df['blosum_scores_fwr1'], bins=20, color='salmon', edgecolor='black', alpha=0.7)
axes[1].set_title('Distribution of BLOSUM Scores for FWR1')
axes[1].set_xlabel('BLOSUM Scores')
axes[1].set_ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()
# Show the plot
plt.show()
```
