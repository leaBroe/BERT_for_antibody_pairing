```python
# env: OAS_paired_env
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/sqlite3_data_for_analysis/evaluate_test_set_by_regions/h2l_div_beam_search_2_epoch_10_lr_1e-4_wd_0.1/fwr1_similarity_blosum_scores.csv')
```

```python
# Create a figure and axis objects for the plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the distribution of similarity scores
axes[0].hist(df['similarity_scores_fwr1'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_title('Distribution of Similarity Scores for FWR1')
axes[0].set_xlabel('Similarity Scores')
axes[0].set_ylabel('Frequency')

# Plot the distribution of blosum scores
axes[1].hist(df['blosum_scores_fwr1'], bins=10, color='salmon', edgecolor='black', alpha=0.7)
axes[1].set_title('Distribution of Blosum Scores for FWR1')
axes[1].set_xlabel('Blosum Scores')
axes[1].set_ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()
# Show the plot
plt.show()
```
