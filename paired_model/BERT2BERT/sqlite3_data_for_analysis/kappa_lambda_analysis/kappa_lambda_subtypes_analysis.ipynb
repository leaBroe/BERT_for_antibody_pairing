````python
# used env: adap_2
import pandas as pd
````

````python
# File paths
file1_path = '/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/logs/full_eval_heavy2light_with_adapters125463.o'
file4_path = '/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/sqlite3_data_for_analysis/kappa_lambda_analysis/kappa_lambda_subtypes_extraction.csv'
additional_file_path = '/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2BERT/sqlite3_data_for_analysis/B-cell_analysis/Btypes_full_paired_test_data_no_dupl.csv'
````

````python
# Parse the first file
with open(file1_path, 'r') as file:
    lines = file.readlines()
````

````python
# Parse the first file
with open(file1_path, 'r') as file:
    lines = file.readlines()

data1 = []
seq_pair = {}
for line in lines:
    line = line.strip()
    if line.startswith('Sequence pair'):
        if seq_pair:
            data1.append(seq_pair)
        seq_pair = {}
        seq_pair['Sequence pair'] = line.split(' ')[2][:-1]
    elif line.startswith('True Sequence:'):
        seq_pair['True Sequence'] = ''.join(line.split(' ')[2:])
    elif line.startswith('Generated Sequence:'):
        seq_pair['Generated Sequence'] = ''.join(line.split(' ')[2:])
    elif line.startswith('BLOSUM Score:'):
        seq_pair['BLOSUM Score'] = float(line.split(' ')[2])
    elif line.startswith('Similarity Percentage:'):
        seq_pair['Similarity Percentage'] = float(line.split(' ')[2][:-1])
    elif line.startswith('Perplexity:'):
        seq_pair['Perplexity'] = float(line.split(' ')[1])

if seq_pair:
    data1.append(seq_pair)

# Convert to DataFrame
df1 = pd.DataFrame(data1)

print(len(df1))
df1
````

````python
# read csv file
df1 = pd.read_csv("extracted_data_alignment_heavy2light_60_epochs_beam_search_127845.csv")
df1
````

````python
# Read the fourth file
df4 = pd.read_csv(file4_path)
df4['sequence_alignment_aa'] = df4['sequence_alignment_aa'].str.replace(' ', '')

print(len(df4))
df4
````

````python
# Read the additional file
df_additional = pd.read_csv(additional_file_path)

df_additional 

# Extract the light sequence after the [SEP] token and remove spaces
df_additional['Light Sequence'] = df_additional['sequence_alignment_heavy_sep_light'].apply(lambda x: x.split('[SEP]')[-1].replace(' ', ''))

# print(len(df_additional))
df_additional
````

````python
# Merge df1 and df4 on True Sequence and sequence_alignment_aa
merged_df = pd.merge(df1, df4, left_on='sequence_alignment_aa_light', right_on='sequence_alignment_aa', how='inner')

#merged_df = pd.merge(merged_df, df_additional, on="sequence_alignment_aa_light", how='inner')
merged_df.drop_duplicates(subset=['sequence_alignment_aa_light'], inplace=True)




print(len(merged_df))  
merged_df
````

````python
# Create the 'subtype' column by extracting the part before the first hyphen in the 'v_family' column
merged_df['subtype'] = merged_df['v_family'].apply(lambda x: x.split('-')[0])

merged_df
````

````python
# group merged_df by v_family and calculate the mean of BLOSUM Score, Similarity Percentage and Perplexity
# also add number of sequences in each group and the min and max values of BLOSUM Score, Similarity Percentage and Perplexity
grouped_df_v_family = merged_df.groupby('subtype').agg({
    'calculated_blosum': ['mean', 'median', 'min', 'max'],
    'calculated_similarity': ['mean', 'median', 'min', 'max'],
    'perplexity': ['mean', 'median', 'min', 'max'],
    'subtype': 'size'  # This adds a count of the number of sequences in each subtype
}).reset_index()

locus_df_v_family = merged_df.groupby('locus').agg({
    'calculated_blosum': ['mean', 'median', 'min', 'max'],
    'calculated_similarity': ['mean', 'median', 'min', 'max'],
    'perplexity': ['mean', 'median', 'min', 'max'],
    'locus': 'size'  # This adds a count of the number of sequences in each subtype
}).reset_index()
# # # group merged_df by lambda_subtype and calculate the median of BLOSUM Score, Similarity Percentage and Perplexity
# grouped_df_v_family_median = merged_df.groupby('v_family').agg({'BLOSUM Score': 'median', 'Similarity Percentage': 'median', 'Perplexity': 'median', 'Sequence pair': 'count'}).reset_index()

# save the grouped_df to a csv file
grouped_df_v_family.to_csv('global_Alignment_kappa_lambda_fewer_subtypes_df.csv', index=False)
locus_df_v_family.to_csv('global_Alignment_kappa_lambda_fewer_locus_df.csv', index=False)

````
