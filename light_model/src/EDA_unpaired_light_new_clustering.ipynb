```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import csv
import statistics
```

# Clustering Bcells_subset_human_unpaired_light
## 100 PIDENT CDRL3 sequences
-k 6 --spaced-kmer-mode 1 --spaced-kmer-pattern 11011101 --sub-mat VTML40.out --gap-open 16 --gap-extend 2

```python
# DATA
paired_cdrh3=pd.read_csv('/ibmm_data2/oas_database/paired_lea_tmp/txt_files_oas_db/Bcells_subset_human_unpaired_light/linclust_new_params/100_pident/cdr3_light_seq_3_rowid_100_clu_rep_idseq',header=None)
print(paired_cdrh3.head(2))

cdr3_aa_len = paired_cdrh3[1].apply(lambda x: len(str(x)))

# Create a figure with 2 subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Boxplot on the first subplot
axs[0].boxplot(cdr3_aa_len)
axs[0].set_ylabel('CDR-L3 length')
axs[0].set_xlabel('CDR-L3')
axs[0].set_title('CDR-L3 - B-Cells subset human unpaired light - 100 PIDENT')
text = f'# centroids seq: {len(paired_cdrh3)}'
axs[0].text(0.23, 0.9, text, transform=axs[0].transAxes, ha='center', va='center',
            bbox=dict(facecolor='white', edgecolor='black'))
text = f'AVG length: {round(statistics.mean(cdr3_aa_len))}'
axs[0].text(0.23, 0.80, text, transform=axs[0].transAxes, ha='center', va='center',
            bbox=dict(facecolor='white', edgecolor='black'))

# KDE plot on the second subplot
sns.histplot(cdr3_aa_len, kde=True, bins=100, ax=axs[1])
axs[1].set_xlabel('CDR-L3 length')
axs[1].set_ylabel('Frequency')
axs[1].set_title('CDR-L3 - B-Cells subset human unpaired light - 100 PIDENT')

plt.tight_layout()

plt.savefig('/ibmm_data2/oas_database/paired_lea_tmp/EDA_plots/light_model/clustering_cdrl3_100_unpaired_light.png')

plt.show()
```
